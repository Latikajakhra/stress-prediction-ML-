# Stress Level Prediction Project

# Introduction
- **Objective**: Predict student stress level using machine learning.
- **Dataset**: 'Student_Behaviour.csv' with features like Gender, Height, Weight, Marks, Daily Studying Time, etc.

## Importing necessary libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
from sklearn.metrics import accuracy_score, classification_report
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load the dataset

```python
df = pd.read_csv('Downloads/archive/Student_Behaviour.csv')
df.head()
```

```python
df.info()
```

## Data Preprocessing

```python
#checking for any missing values
df.isnull().sum()
```

```python
df.columns
```

```python
# Function to convert categorical time strings to numerical values
def convert_travel_time(time_str):
    if time_str == '0 - 30 minutes':
        return 15
    elif time_str == '30 - 60 minutes':
        return 45
    elif time_str == '1 - 1.30 hour':
        return 75
    elif time_str == '1.30 - 2 hour':
        return 105
    elif time_str == '2 - 2.30 hour':
        return 135
    elif time_str == '2.30 - 3 hour':
        return 165
    elif time_str == 'more than 3 hour':
        return 270
    else:
        return 0
def convert_dailystuding_time(time_str):
    if time_str == '0 - 30 minute':
        return 15
    elif time_str == '30 - 60 minute':
        return 45
    elif time_str == '1 - 2 Hour':
        return 90
    elif time_str == '2 - 3 hour':
        return 150
    elif time_str == '3 - 4 hour':
        return 210
    elif time_str == 'More Than 4 hour':
        return 270
    else:
        return 0

def convert_percentage_to_float(percentage_str):
    try:
        return float(percentage_str.strip('%'))
    except ValueError:
        return 0.0
        
def convert_socialmedai_time(time_str):
    if time_str == '0 Minute':
        return 0
    elif time_str == '1 - 30 Minute':
        return 15
    elif time_str == '30 - 60 Minute':
        return 45
    elif time_str == '1 - 1.30 hour':
        return 75
    elif time_str == '1.30 - 2 hour':
        return 105
    elif time_str == 'More than 2 hour':
        return 210
    else:
        return 0
# Apply the conversion functions to relevant columns
df['Travelling Time '] = df['Travelling Time '].apply(convert_travel_time)
df['daily studing time'] = df['daily studing time'].apply(convert_dailystuding_time)
df['willingness to pursue a career based on their degree  '] = df['willingness to pursue a career based on their degree  '].apply(convert_percentage_to_float)
df['social medai & video'] = df['social medai & video'].apply(convert_socialmedai_time)


# Encode categorical variables

    # Label encoding 
label_encoder = LabelEncoder()
binary_columns = ['Gender', 'Do you like your degree?', 'part-time job']
for col in binary_columns:
    df[col] = label_encoder.fit_transform(df[col])
    
    # Ordinal encoding 
ordinal_encoder = OrdinalEncoder(categories=[['Fabulous', 'good', 'Bad', 'Awful']])
df['Financial Status'] = ordinal_encoder.fit_transform(df[['Financial Status']])
ordinal_encoder_study_time = OrdinalEncoder(categories=[['Morning', 'Anytime', 'Night']])
df['prefer to study in'] = ordinal_encoder_study_time.fit_transform(df[['prefer to study in']])
ordinal_encoder_stress = OrdinalEncoder(categories=[['fabulous', 'Good', 'Bad', 'Awful']])
df['Stress Level '] = ordinal_encoder_stress.fit_transform(df[['Stress Level ']])
```

```python
df.head()
```

## Model Training

```python
# Define features and target variable
X = df[['Gender', 'Height(CM)', 'Weight(KG)', 'college mark', '12th Mark', 'daily studing time', 'prefer to study in', 'salary expectation', 'Do you like your degree?', 'willingness to pursue a career based on their degree  ', 'social medai & video', 'Travelling Time ', 'Financial Status', 'part-time job']]
y = df['Stress Level ']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the Random Forest classifier with class weights
rf_clf = RandomForestClassifier(random_state=42, class_weight='balanced')
rf_clf.fit(X_train, y_train)
```

## Model Evaluation

```python
# Evaluate the model
y_pred = rf_clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
print(classification_report(y_test, y_pred, zero_division=1))

# Save the model
joblib.dump(rf_clf, 'trained_model.pkl')
```

## Stress Prediction

```python
def predict_stress_level():
    gender = int(input("Enter your gender (0 for male, 1 for female): "))
    height = float(input("Enter your height in centimeters: "))
    weight = float(input("Enter your weight in kilograms: "))
    college_mark = float(input("Enter your college mark percentage: "))
    twelfth_mark = float(input("Enter your 12th mark percentage: "))
    daily_studing_time = convert_dailystuding_time(input("Enter your daily studying time (e.g., '1 - 2 Hour'): "))
    prefer_to_study_in = int(input("Enter your preference to study in (0 for Morning, 1 for Anytime, 2 for Night): "))
    salary_expectation = float(input("Enter your salary expectation: "))
    do_you_like_your_degree = int(input("Do you like your degree? (0 for No, 1 for Yes): "))
    willingness_to_pursue = convert_percentage_to_float(input("Enter your willingness to pursue a career based on your degree (percentage): "))
    social_media_time = convert_socialmedai_time(input("Enter your social media & video time (e.g., '1 - 1.30 hour'): "))
    travelling_time = convert_travel_time(input("Enter your travelling time (e.g., '1 - 1.30 hour'): "))
    financial_status = int(input("Enter your financial status (0 for Fabulous, 1 for Good, 2 for Bad, 3 for Awful): "))
    part_time_job = int(input("Do you have a part-time job? (0 for No, 1 for Yes): "))
    
        # Convert input to a DataFrame with the same columns as the training data
    user_input = pd.DataFrame([{
        'Gender': gender,
        'Height(CM)': height,
        'Weight(KG)': weight,
        'college mark': college_mark,
        '12th Mark': twelfth_mark,
        'daily studing time': daily_studing_time,
        'prefer to study in': prefer_to_study_in,
        'salary expectation': salary_expectation,
        'Do you like your degree?': do_you_like_your_degree,
        'willingness to pursue a career based on their degree  ': willingness_to_pursue,
        'social medai & video': social_media_time,
        'Travelling Time ': travelling_time,
        'Financial Status': financial_status,
        'part-time job': part_time_job
    }])
    # Use the trained model to make predictions
    prediction = rf_clf.predict(user_input)

    # Convert the prediction to a meaningful result (e.g., stress level category)
    result = ordinal_encoder_stress.inverse_transform(prediction.reshape(-1, 1))[0][0]
    print(f"Based on your input, your predicted stress level is: {result}")


# Test the prediction function
predict_stress_level()

```
